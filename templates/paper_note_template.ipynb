```python
import numpy as np
import matplotlib.pyplot as plt

# Windows 下 matplotlib 中文显示
plt.rcParams['font.sans-serif'] = ['SimHei', 'Microsoft YaHei']
plt.rcParams['axes.unicode_minus'] = False

rng = np.random.default_rng(42)  # 固定随机种子，保证结果可复现
```

```python
# ===== 实验代码 =====
# 示例：SGD 在强凸二次函数上的收敛，验证 O(1/T) / O(1/sqrt(T)) 速率

def run_sgd(eta=0.01, T=2000, d=10, noise=1.0):
    A = np.diag(np.linspace(1.0, 10.0, d))   # 条件数 10 的二次目标 f(x) = x^T A x / 2
    x = rng.standard_normal(d)
    losses = []
    for t in range(T):
        g = A @ x + noise * rng.standard_normal(d)
        x = x - eta * g
        losses.append(0.5 * x @ A @ x)
    return np.array(losses)

losses = run_sgd()

plt.figure(figsize=(6, 4))
plt.loglog(losses, label=r'SGD, $\eta=0.01$')
plt.xlabel('迭代次数 $t$')
plt.ylabel('损失 $f(x_t)$')
plt.title('示例实验（替换成你自己的）')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```
